# **Q1: What is a Support Vector Machine (SVM)**

-> A Support Vector Machine (SVM) is a powerful supervised learning algorithm used in machine learning for solving complex classification, regression, and outlier detection problems. SVMs are particularly effective in scenarios where the data is not linearly separable, as they can transform the data into higher-dimensional spaces to find optimal decision boundaries.

Key Features of SVMs:
- Decision Boundary (Hyperplane): SVMs aim to find the best hyperplane that maximizes the margin between different classes. This hyperplane acts as a decision boundary that separates classes effectively.

Types of SVMs:

- Linear SVM: Uses a linear kernel for linearly separable data, providing a straightforward decision boundary.

- Nonlinear SVM: Employs kernel functions (e.g., polynomial, Gaussian) to map data into higher-dimensional spaces for non-linear separation.

- Support Vector Regression (SVR): Designed for regression tasks, focusing on modeling continuous output values.

- One-Class SVM: Used for outlier detection by identifying data points that do not belong to a well-defined class.

- Multiclass SVM: Handles multiple classes by training multiple binary classifiers.

How SVMs Work:
- Data Transformation: SVMs can transform data into higher-dimensional spaces using kernel functions, allowing for linear separation of non-linear data.

- Hyperplane Selection: The algorithm selects the hyperplane with the maximum margin, ensuring robust classification.

- Support Vectors: The data points closest to the hyperplane are called support vectors, as they define the boundary.

Applications of SVMs:

SVMs are widely used in various fields, including healthcare, natural language processing, image recognition, and signal processing, due to their ability to handle complex data and provide robust classification and regression capabilities



# **Q2. What is the difference between Hard Margin and Soft Margin SVM?**

-> The primary difference between Hard Margin SVM and Soft Margin SVM lies in their approach to handling data separability and misclassifications:

Hard Margin SVM
- Objective: Maximize the margin between classes without allowing any misclassifications.

- Assumption: Data must be linearly separable.

- Limitations: Fails when data contains outliers or is not linearly separable. It is sensitive to noise and outliers, which can lead to overfitting.

- Constraints: All data points must be correctly classified and lie at least at a distance of 1 from the hyperplane.

Soft Margin SVM:

- Objective: Balance between maximizing the margin and minimizing misclassifications by introducing slack variables.

- Assumption: Can handle non-linearly separable data or noisy data.

- Advantages: More robust to outliers and noise compared to hard margin SVM. It allows for some misclassifications to improve generalization.

- Constraints: Slack variables are used to allow data points to be misclassified, controlled by the regularization parameter C.


# **Q3. What is the mathematical intuition behind SVM**

-> The mathematical intuition behind Support Vector Machines (SVMs) involves understanding how they optimize the decision boundary (hyperplane) to classify data effectively. Here's a breakdown of the key mathematical concepts:

1. Optimization Problem
- SVMs are formulated as an optimization problem, where the goal is to find the hyperplane that maximizes the margin between classes. This is achieved by solving a quadratic programming problem.

2. Hyperplane Equation
- The hyperplane in SVMs is defined by the equation:

  w⋅x+b=0

where:

- w is the weight vector perpendicular to the hyperplane.

- x is the input vector.

- b is the bias term.

3. Margin Maximization
The margin is the distance between the hyperplane and the nearest data points (support vectors). Maximizing the margin ensures that the model generalizes well to unseen data. The margin is given by:
Margin=
2/∥w∥


where

∥w∥ is the norm of the weight vector.

4. Kernel Functions
- For non-linearly separable data, SVMs use kernel functions to map the data into higher-dimensional spaces where it becomes linearly separable.

5. Soft Margin and Slack Variables
- In cases where data is not linearly separable, SVMs introduce slack variables (ξi) to allow for misclassifications.

# **Q5. What are Support Vectors in SVM**
-> Support vectors in Support Vector Machines (SVMs) are the critical data points that lie closest to the decision boundary (or hyperplane) and directly influence its position and orientation. These points play a pivotal role in defining the margin between different classes. Here's a detailed explanation:

Definition of Support Vectors
Support vectors are the subset of training data points that satisfy the following conditions:

They lie on or inside the margin boundaries.

Their corresponding Lagrange multipliers (αi) are non-zero in the dual optimization problem.

These points are essential because they determine the optimal hyperplane for classification.

Role of Support Vectors
1. Defining the Decision Boundary
The decision boundary (hyperplane) is determined using only the support vectors, not the entire dataset.

2. Maximizing the Margin
- Support vectors are located at or near the margin boundaries, ensuring that the margin between classes is maximized while maintaining correct classification.

3. Sparse Representation
- Since only a small subset of data points (support vectors) influence the hyperplane, SVM achieves computational efficiency and scalability.


# **Q6. What is a Support Vector Classifier (SVC)?**

-> A Support Vector Classifier (SVC) is a type of machine learning model used for classification tasks. It is a supervised learning algorithm that aims to find the optimal decision boundary (hyperplane) that maximally separates different classes in the data. Here's a detailed overview:

### Key Features of SVC

Decision Boundary
- SVC seeks a hyperplane that maximizes the margin between classes, ensuring robust classification. This hyperplane is defined by the equation w (raiese to the power T) x+b=0, where w is the weight vector, x is the feature vector, and b is the bias term.

Max-margin Principle
- The goal is to maximize the distance between the hyperplane and the nearest data points of each class. This principle enhances the model's generalization ability by reducing overfitting.

Support Vectors
- The decision boundary is determined by a subset of data points called support vectors. These are the points closest to the hyperplane and have non-zero Lagrange multipliers (αi) in the dual optimization problem.

Kernel Methods
- SVC can handle non-linearly separable data by using kernel functions to transform the data into a higher-dimensional space where a linear separation is possible. Common kernels include polynomial, radial basis function (RBF), and sigmoid.

Types of SVC

- Linear SVC: Used for linearly separable data, where a straight line can separate classes.

- Non-linear SVC: Employs kernel functions for non-linearly separable data, allowing complex decision boundaries.

- Multiclass SVC: Can be extended to classify more than two classes using techniques like One-vs-One (OvO) or One-vs-All (OvA).


Advantages
- Robustness: SVCs are resilient to noisy data and outliers.

- High-dimensional Data: Effective in high-dimensional spaces.

- Complex Decision Boundaries: Can create complex boundaries using kernel methods.


Applications
- SVCs are widely used in fields like image recognition, text classification, and bioinformatics, where they provide robust classification capabilities

# **Q7. What is a Support Vector Regressor (SVR)**

-> A Support Vector Regressor (SVR) is an extension of Support Vector Machines (SVMs) designed for regression tasks. It aims to find a function that approximates the relationship between input data and continuous output variables. Here's a detailed overview:

Key Features of SVR

1. Epsilon-Insensitive Loss
- SVR uses an epsilon-insensitive loss function, which allows for a margin of error (epsilon tube) around the predicted values. Data points within this tube are considered correctly predicted and do not contribute to the loss. Points outside the tube are penalized proportionally to their distance from the tube.

2. Hyperplane in Regression
- Unlike SVM classification, where the hyperplane separates classes, SVR uses a hyperplane to predict continuous values. The goal is to find a hyperplane that minimizes the error between predicted and actual values while being as flat as possible.

3. Kernel Methods
- SVR can handle non-linear relationships by employing kernel functions to transform data into higher-dimensional spaces, similar to SVMs. Common kernels include polynomial, radial basis function (RBF), and sigmoid.

4. Support Vectors
- In SVR, support vectors are data points that lie outside the epsilon tube. These points have non-zero Lagrange multipliers and directly influence the position of the regression line.


Mathematical Formulation

- The SVR problem can be formulated as follows:

- Objective: Minimize the sum of the absolute deviations between predicted and actual values for points outside the epsilon tube.

- Constraints: Ensure that the predicted values for points within the epsilon tube are close enough to the actual values.

Advantages

- Robustness to Noise: SVR is robust to noisy data due to its epsilon-insensitive loss function.

- Handling Non-Linear Relationships: Effective in modeling complex, non-linear relationships using kernel functions.

- Flexibility: Can be used for both linear and non-linear regression tasks.

Applications

- SVR is widely used in applications such as time series forecasting, stock price prediction, and energy demand forecasting, where accurate continuous predictions are crucial.

Comparison with Linear Regression

- SVR vs. Linear Regression: SVR does not require specifying the relationship between variables explicitly, unlike linear regression. It automatically determines these relationships.

- Robustness: SVR is more robust to outliers compared to traditional linear regression due to its epsilon-insensitive loss function.

# **Q8. What is the Kernel Trick in SVM?**

-> The Kernel Trick is a fundamental technique used in Support Vector Machines (SVMs) to handle non-linearly separable data. It allows SVMs to operate in higher-dimensional spaces without explicitly computing the transformations, enabling the creation of complex, non-linear decision boundaries.

How the Kernel Trick Works

1. Mapping to Higher Dimensions
- Instead of directly transforming data into a higher-dimensional space, the kernel trick uses a kernel function to compute the inner products between data points in the new space. This avoids the need for explicit mapping.

2. Dual Problem Formulation
- The kernel trick is applied in the dual formulation of SVMs, where the goal is to maximize the margin between classes. The dual problem involves optimizing over the Lagrange multipliers (αi), using kernel functions to compute inner products between data points in the higher-dimensional space.

3. Kernel Functions

Common kernel functions include:

- Linear Kernel
- Polynomial Kernel
- Radial Basis Function (RBF) Kernel

Advantages of the Kernel Trick:

- Efficiency: Avoids explicit computation of high-dimensional transformations, reducing computational complexity.

- Flexibility: Enables SVMs to handle non-linear relationships by creating complex decision boundaries.

- Robustness: Enhances the robustness of SVMs by allowing them to operate effectively in high-dimensional spaces.

Applications
- The kernel trick is crucial in various applications, including text classification, image recognition, and bioinformatics, where complex data relationships need to be modeled efficiently.

Mathematical Basis
- The kernel trick relies on Mercer's theorem, which ensures that a kernel function can be represented as an inner product in a higher-dimensional space. This allows SVMs to leverage the power of kernel functions without explicit mapping.

# **Q10. What is the effect of the C parameter in SVM**
-> The C parameter in Support Vector Machines (SVMs) plays a crucial role in controlling the trade-off between the model's ability to correctly classify training examples and the maximization of the margin. Here's how it affects SVM performance:

Role of C Parameter:

- Misclassification Penalty: The C parameter determines the penalty for misclassified data points during training. A higher value of C increases the penalty, while a lower value decreases it.

Margin vs. Accuracy:

- Higher C: Prioritizes correct classification of all training examples, potentially leading to a narrower margin. This can result in overfitting if the data is noisy or not well-separated.

- Lower C: Emphasizes maximizing the margin, allowing more misclassifications. This can help prevent overfitting and improve generalization on unseen data.

Effects on SVM Performance:

- Overfitting vs. Underfitting:

-  High C: Increases the risk of overfitting by fitting the model too closely to the training data.

-  Low C: May lead to underfitting if the model becomes too simplistic and fails to capture underlying patterns.

- Robustness to Noise: Lower C values can be beneficial in noisy datasets by allowing for some misclassifications and focusing on a wider margin.

Choosing the Optimal C

- Selecting the optimal C value is crucial for achieving a well-performing SVM model. Techniques like grid search with cross-validation are commonly used to find the best C value based on the dataset and desired trade-offs.

# **Q11.  What is the role of the Gamma parameter in RBF Kernel SVM**
-> The Gamma parameter in the Radial Basis Function (RBF) kernel of a Support Vector Machine (SVM) plays a crucial role in determining the complexity of the decision boundary. Here's how it influences SVM performance:

Role of Gamma:

- Spread of the Kernel: Gamma controls the spread or width of the RBF kernel. A smaller gamma results in a larger spread, meaning the influence of a single training example reaches farther. Conversely, a larger gamma leads to a smaller spread, where the influence is more localized.

Decision Boundary Complexity:


  -Low Gamma: Produces smoother decision boundaries, which can lead to underfitting if too simple.

  -High Gamma: Results in more complex, wiggly decision boundaries, which can lead to overfitting if too complex.

- Trade-off Between Bias and Variance: Gamma controls the trade-off between model bias (underfitting) and variance (overfitting). A high gamma increases the model's ability to fit the training data closely but risks overfitting.

Choosing the Optimal Gamma:
- Selecting the right gamma value is essential for achieving a well-performing SVM model. Techniques like cross-validation are often used to find the optimal gamma value based on the dataset and desired performance metrics

# **Q12. What is the Naïve Bayes classifier, and why is it called "Naïve"**

-> The Naïve Bayes classifier is a supervised machine learning algorithm used for classification tasks. It is based on Bayes' Theorem, which calculates the probability of an event given prior knowledge of related conditions. Naïve Bayes is particularly effective in text classification tasks such as spam filtering and sentiment analysis due to its simplicity and efficiency in handling high-dimensional data.

###Why is it Called "Naïve"?

- The Naïve Bayes classifier is called "naïve" because it makes a strong assumption of conditional independence among features. This means that the presence of one feature does not affect the presence of another feature, given the class. In reality, features often correlate, but this simplification allows for quick and efficient calculations, making it suitable for large datasets and real-time applications.

Key Assumptions and Characteristics
- Assumption of Conditional Independence: Features are assumed to be independent of each other, given the class.

- Probabilistic Nature: Uses Bayes' Theorem to calculate posterior probabilities for classification.

- Simplicity and Efficiency: Easy to implement and fast in both training and prediction phases.

- Scalability: Handles high-dimensional data effectively.

- Despite its simplistic assumptions, Naïve Bayes often performs well in practice, especially in scenarios where speed and simplicity are crucial.

# **Q13. What is Bayes’ Theorem**
-> Bayes' Theorem is a mathematical formula that describes how to update the probability of a hypothesis based on new evidence. It is named after the 18th-century British mathematician Thomas Bayes and is foundational in the field of Bayesian statistics.

Mathematical Formula:

- Bayes' Theorem can be expressed mathematically as:

  P(A∣B)=P(B∣A)⋅P(A)/P(B)

Where:
•	P(A∣B)P(A∣B) is the posterior probability: the probability of event AA occurring given that BB is true.
•	P(B∣A)P(B∣A) is the likelihood: the probability of event BB occurring given that AA is true.
•	P(A)P(A) is the prior probability: the initial probability of event AA before observing BB.
•	P(B)P(B) is the marginal probability: the total probability of event BB.


Applications
- Bayes' Theorem has widespread applications across various fields, including:

- Medicine: To evaluate the accuracy of diagnostic tests by considering both the prevalence of diseases and test accuracy.

- Finance: For risk assessment and decision-making based on prior knowledge and new data.

- Machine Learning: In algorithms such as Naïve Bayes classifiers for text classification and spam detection.

Significance:

- Bayes' Theorem allows for updating predictions or beliefs in light of new evidence, making it a powerful tool for statistical inference and decision-making. Its ability to incorporate prior knowledge into probability assessments makes it particularly valuable in uncertain environments.

# Q20.  How does Laplace Smoothing help in Naïve Bayes? **

-> Laplace smoothing is a technique that addresses the problem of zero probabilities in Naïve Bayes classifiers. Here's how it works and why it's important:

### Problem of Zero Probabilities:
In Naïve Bayes, probabilities of features are calculated based on their frequency in the dataset. If a feature (or word, in text classification) has never been observed in the training set for a specific class, its probability becomes zero. Since Naïve Bayes multiplies probabilities, this zero probability causes the entire prediction for that class to collapse, regardless of other evidence.

### How Laplace Smoothing Helps:
1. **Adding a Small Constant**:
   - Laplace smoothing adds a small positive constant (usually 1) to the numerator of the probability formula for each feature.
   - This ensures that no probability is ever zero, even for unseen features.

2. **Modified Probability Formula**:
   - Without smoothing:  
     Probability = (Number of times the word appears in spam) ÷ (Total number of words in spam)
   - With Laplace smoothing:  
     Probability = (Number of times the word appears in spam + 1) ÷ (Total number of words in spam + Total unique words)

3. **Prevents Bias Towards Seen Features**:
   - By assigning a small probability to unseen features, Laplace smoothing allows the model to handle unseen data better and prevents it from being overly confident in its predictions.

### Real-World Impact:
In applications like spam detection or text classification, it's common to encounter new words during testing that weren’t present in the training set. Laplace smoothing ensures the model can still make reasonable predictions in such cases.



```python
'''21. Write a Python program to train an SVM Classifier on the Iris dataset and evaluate accuracy'''

from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data  # Features
y = iris.target  # Labels

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create an SVM classifier
svm_classifier = SVC(kernel='linear', random_state=42)

# Train the classifier
svm_classifier.fit(X_train, y_train)

# Make predictions on the test set
y_pred = svm_classifier.predict(X_test)

# Evaluate the accuracy of the classifier
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of the SVM classifier: {accuracy:.2f}")
```

```python
'''22.  Write a Python program to train two SVM classifiers with Linear and RBF kernels on the Wine dataset, then compare their accuracies: '''

from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score


wine = datasets.load_wine()
X = wine.data
y = wine.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create SVM classifiers with Linear and RBF kernels
svm_linear = SVC(kernel='linear', random_state=42)
svm_rbf = SVC(kernel='rbf', random_state=42)

# Train the classifiers
svm_linear.fit(X_train, y_train)
svm_rbf.fit(X_train, y_train)

# Make predictions on the test set
y_pred_linear = svm_linear.predict(X_test)
y_pred_rbf = svm_rbf.predict(X_test)

# Evaluate and compare accuracies
accuracy_linear = accuracy_score(y_test, y_pred_linear)
accuracy_rbf = accuracy_score(y_test, y_pred_rbf)

print(f"Accuracy with Linear kernel: {accuracy_linear:.2f}")
print(f"Accuracy with RBF kernel: {accuracy_rbf:.2f}")

# Compare results
if accuracy_linear > accuracy_rbf:
    print("The Linear kernel performed better.")
elif accuracy_rbf > accuracy_linear:
    print("The RBF kernel performed better.")
else:
    print("Both kernels performed equally well.")
```

```python
'''23.  Write a Python program to train an SVM Regressor (SVR) on a housing dataset and evaluate it using Mean Squared Error (MSE): '''


from sklearn import datasets
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Load the housing dataset (Boston Housing Dataset)
from sklearn.datasets import fetch_california_housing
housing_data = fetch_california_housing()
X = housing_data.data
y = housing_data.target  # Target variable (house prices)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create an SVR model with the RBF kernel
svr_model = SVR(kernel='rbf', C=1.0, epsilon=0.2)

# Train the SVR model
svr_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = svr_model.predict(X_test)

# Evaluate the model using Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error (MSE) of the SVR model: {mse:.2f}")
```

```python
'''24. Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision boundary '''


import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import make_classification
from sklearn.preprocessing import StandardScaler

# Generate a synthetic 2D dataset
X, y = make_classification(
    n_samples=100, n_features=2, n_classes=2, n_clusters_per_class=1, n_redundant=0, random_state=42
)

# Standardize the features for better performance
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Create an SVM classifier with a polynomial kernel
svm_poly = SVC(kernel='poly', degree=3, C=1, random_state=42)

# Train the SVM classifier
svm_poly.fit(X, y)

# Visualize the decision boundary
# Create a mesh grid
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500), np.linspace(y_min, y_max, 500))

# Predict classes for each point in the mesh grid
Z = svm_poly.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundary
plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.Paired)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap=plt.cm.Paired)
plt.title("SVM with Polynomial Kernel (Degree=3)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
'''27. Write a Python program to train an SVM Classifier with different C values and compare the decision boundaries visually'''

import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import make_classification
from sklearn.preprocessing import StandardScaler

# Generate a synthetic 2D dataset for visualization
X, y = make_classification(
    n_samples=100, n_features=2, n_classes=2, n_clusters_per_class=1, n_redundant=0, random_state=42
)

# Standardize the features for better performance
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Define different values of C to test
C_values = [0.1, 1, 10]

# Create a function to plot decision boundaries
def plot_decision_boundaries(X, y, model, ax, title):
    # Create a mesh grid for plotting
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500), np.linspace(y_min, y_max, 500))

    # Predict the decision boundary
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plot the decision boundary and data points
    ax.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.Paired)
    ax.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap=plt.cm.Paired)
    ax.set_title(title)

# Create subplots to compare results
fig, axes = plt.subplots(1, len(C_values), figsize=(15, 5), sharey=True)

for i, C in enumerate(C_values):
    # Train an SVM classifier with a specific C value
    svm = SVC(kernel='linear', C=C, random_state=42)
    svm.fit(X, y)

    # Plot the decision boundary
    plot_decision_boundaries(X, y, svm, axes[i], title=f"SVM with C={C}")

# Display the plots
plt.tight_layout()
plt.show()
```

```python
'''35.  Write a Python program to perform feature selection before training a Naïve Bayes classifier and compare results'''


from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = load_iris()
X = iris.data  # Features
y = iris.target  # Labels

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train a Naïve Bayes classifier without feature selection
nb_classifier = GaussianNB()
nb_classifier.fit(X_train, y_train)
y_pred = nb_classifier.predict(X_test)
accuracy_without_fs = accuracy_score(y_test, y_pred)
print(f"Accuracy without feature selection: {accuracy_without_fs:.2f}")

# Perform feature selection using SelectKBest with Chi-Squared test
k = 2  # Number of features to select
selector = SelectKBest(score_func=chi2, k=k)
X_train_selected = selector.fit_transform(X_train, y_train)
X_test_selected = selector.transform(X_test)

# Train a Naïve Bayes classifier with selected features
nb_classifier_selected = GaussianNB()
nb_classifier_selected.fit(X_train_selected, y_train)
y_pred_selected = nb_classifier_selected.predict(X_test_selected)
accuracy_with_fs = accuracy_score(y_test, y_pred_selected)
print(f"Accuracy with feature selection (top {k} features): {accuracy_with_fs:.2f}")

# Compare the results
if accuracy_with_fs > accuracy_without_fs:
    print("Feature selection improved the accuracy.")
elif accuracy_with_fs < accuracy_without_fs:
    print("Feature selection reduced the accuracy.")
else:
    print("Feature selection had no effect on the accuracy.")
```

```python
'''40. Write a Python program to perform Recursive Feature Elimination (RFE) before training an SVM Classifier and compare accuracy'''


from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.feature_selection import RFE
from sklearn.metrics import accuracy_score

# Load the Breast Cancer dataset
data = load_breast_cancer()
X = data.data  # Features
y = data.target  # Labels

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train an SVM classifier without feature selection
svm_classifier = SVC(kernel='linear', random_state=42)
svm_classifier.fit(X_train, y_train)
y_pred = svm_classifier.predict(X_test)
accuracy_without_fs = accuracy_score(y_test, y_pred)
print(f"Accuracy without feature selection: {accuracy_without_fs:.2f}")

# Perform Recursive Feature Elimination (RFE) to select top features
num_features_to_select = 10  # Number of features to keep
rfe_selector = RFE(estimator=SVC(kernel='linear', random_state=42), n_features_to_select=num_features_to_select)
rfe_selector.fit(X_train, y_train)

# Transform the dataset to include only the selected features
X_train_selected = rfe_selector.transform(X_train)
X_test_selected = rfe_selector.transform(X_test)

# Train an SVM classifier with selected features
svm_classifier_selected = SVC(kernel='linear', random_state=42)
svm_classifier_selected.fit(X_train_selected, y_train)
y_pred_selected = svm_classifier_selected.predict(X_test_selected)
accuracy_with_fs = accuracy_score(y_test, y_pred_selected)
print(f"Accuracy with feature selection (top {num_features_to_select} features): {accuracy_with_fs:.2f}")

# Compare results
if accuracy_with_fs > accuracy_without_fs:
    print("Feature selection improved the accuracy.")
elif accuracy_with_fs < accuracy_without_fs:
    print("Feature selection reduced the accuracy.")
else:
    print("Feature selection had no effect on the accuracy.")
```

```python
''' 46. Write a Python program to train an SVM Classifier and visualize the Precision-Recall Curve.'''


import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import precision_recall_curve, auc

# Load the Breast Cancer dataset
data = load_breast_cancer()
X = data.data  # Features
y = data.target  # Labels (binary: 0 and 1)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train an SVM classifier
svm_classifier = SVC(kernel='linear', probability=True, random_state=42)
svm_classifier.fit(X_train, y_train)

# Predict probabilities for the positive class
y_scores = svm_classifier.predict_proba(X_test)[:, 1]

# Compute Precision-Recall values
precision, recall, thresholds = precision_recall_curve(y_test, y_scores)

# Compute the area under the Precision-Recall curve (AUC)
pr_auc = auc(recall, precision)

# Visualize the Precision-Recall Curve
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, label=f'Precision-Recall Curve (AUC = {pr_auc:.2f})')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc='best')
plt.grid()
plt.show()
```
